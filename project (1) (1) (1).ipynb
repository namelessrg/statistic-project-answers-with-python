```python
import pandas as pd
import matplotlib.pyplot as plt
import openpyxl
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_excel(io = r'C:\Users\syldz\Downloads\Employee_Data.xlsx', dtype = {'EEID':object, 'Full Name':object,'Job Title':object,'Department':object,'Business Unit':object,'Gender':object,'Ethnicity':object, 'Age':int, 'Annual Salary':int, 'Bonus':int, 'Country':object, 'City':object}, parse_dates = ['Hire Date', 'Exit Date'])
print (data)
```

```python
mean_annual_salary = data['Annual Salary'].mean()
print("The Mean Value of the Annual Salary:", mean_annual_salary)
```

```python
plt.scatter(data.index , data['Annual Salary'])
plt.xlabel('Index of Employees')
plt.ylabel('Annual Salary')
plt.show()
```

```python

data['Hire Date'] = pd.to_datetime(data['Hire Date'], errors='coerce')

data['Year'] = data['Hire Date'].dt.year

plt.figure(figsize=(8, 6))
sns.histplot(data['Year'], bins=range(min(data['Year']), max(data['Year']) + 1), discrete=True)
plt.xlabel('Year of Hire')
plt.ylabel('Number of Employees')
plt.title('Number of Employees Hired Each Year')

# Display the plot
plt.show()
```

```python
plt.hist(data['Age'], bins=range(min(data['Age']), max(data['Age'])+1))
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.title('Histogram of Employees by Age')
plt.show()
```

```python
plt.boxplot(data['Annual Salary'])
plt.ylabel('Salary')
plt.title('Box Plot of the Salary Data')
plt.show()
```

```python
data_male = data[data['Gender'] == 'Male']
data_female = data[data['Gender'] == 'Female']

fig, axs = plt.subplots(1,2, figsize = (10,5))
sns.boxplot( x = 'Gender' , y = 'Annual Salary', data=data_male, ax = axs[0])
axs[0].set_title('Salary of Males')

sns.boxplot(x = 'Gender' , y = 'Annual Salary', data=data_female, ax = axs[1])
axs[1].set_title('Salary of Females')

plt.tight_layout()
plt.show()
```

```python
data_male = data[data['Gender'] == 'Male']
data_female = data[data['Gender'] == 'Female']
fig, axs = plt.subplots(1,2, figsize = (10,5))
sns.stripplot( x = 'Gender' , y = 'Annual Salary', data=data_male, ax = axs[0])
axs[0].set_title('Salary of Males')
sns.stripplot(x = 'Gender' , y = 'Annual Salary', data=data_female, ax = axs[1])
axs[1].set_title('Salary of Females')
plt.tight_layout()
plt.show()
```

```python
data_male = data[data['Gender'] == 'Male']
data_female = data[data['Gender'] == 'Female']
fig, axs = plt.subplots(1,2, figsize = (10,5))
sns.swarmplot( x = 'Gender' , y = 'Annual Salary', data=data_male, ax = axs[0])
axs[0].set_title('Salary of Males')
sns.swarmplot(x = 'Gender' , y = 'Annual Salary', data=data_female, ax = axs[1])
axs[1].set_title('Salary of Females')
plt.tight_layout()
plt.show()
```

```python
mean_annual_salary = data['Annual Salary'].mean()
std_annual_salary = data['Annual Salary'].std()
print("Mean Annual Salary:", mean_annual_salary)
print("Standard Deviation of Annual Salary:", std_annual_salary)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Ethnicity', data=data)
plt.xlabel('Ethnicity')
plt.ylabel('Number of Employees')
plt.title('Number of Employees by Ethnicity')
plt.xticks(rotation=45)
plt.show()
```

```python
ethnicity_salary_stats = data.groupby('Ethnicity')['Annual Salary'].agg(['mean', 'std'])
print(ethnicity_salary_stats)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Country', data=data)
plt.xlabel('Country')
plt.ylabel('Number of Employees')
plt.title('Number of Employees by Country')
plt.xticks(rotation=45)
plt.show()
```

```python
country_stats = data.groupby('Country')['Annual Salary'].agg(['mean', 'std'])
print("Mean and Standard Deviation of salary data by country:")
print(country_stats)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Department', data=data)
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.title('Number of Employees by Department')
plt.xticks(rotation=45)
plt.show()
```

```python
department_salary_stats = data.groupby('Department')['Annual Salary'].agg(['mean', 'std'])
print(department_salary_stats)
```

```python
sample = data.sample(n=100, random_state=1)
sample_mean = sample['Annual Salary'].mean()
sample_std = sample['Annual Salary'].std()
print("Sample Mean Annual Salary:", sample_mean)
print("Sample Standard Deviation of Annual Salary:", sample_std)
```

```python
sample_means = []
for _ in range(10000):
    sample = data.sample(n=100)
    sample_means.append(sample['Annual Salary'].mean())

plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=50, edgecolor='k')
plt.xlabel('Sample Mean of Annual Salary')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means (n=100, 10,000 samples)')
plt.show()
```
